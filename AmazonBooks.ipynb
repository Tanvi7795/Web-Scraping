```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
ser = Service("C:\\Users\\Maanvi Gupta\\OneDrive\\Documents\\chromedriver.exe")
op = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser, options=op)
```

```python
def get_data(pageno):
    url=('https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'?ie=UTF8&pg='+str(pageno))
    driver.get(url)
    soup=BeautifulSoup(driver.page_source, 'html.parser')
    table=[]
    
    for book in soup.find_all('li', class_='zg-item-immersion'):
        book_details=[]
        hyperlink=book.find('a', class_='a-link-normal')['href']
        hyperlink=str('https://amazon.in')+hyperlink
        driver.get(hyperlink)
        booksoup=BeautifulSoup(driver.page_source, 'html.parser')
        bullet_list=booksoup.find('ul', class_='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list')
        bullet=bullet_list.find_all('span', class_='')
        try:
            publisher=bullet[0].text
        except Exception as e:
            publisher=None
        book_details.append(publisher)
        try:
            language=bullet[1].text
        except Exception as e:
            language=None
        book_details.append(language)
        try:
            paperback=bullet[2].text
        except Exception as e:
            paperback=None
        book_details.append(paperback)
        try:
            ISBN_10=bullet[3].text
        except Exception as e:
            ISBN_10=None
        book_details.append(ISBN_10)
        try:
            ISBN_13=bullet[4].text
        except Exception as e:
            ISBN_13=None
        book_details.append(ISBN_13)
        try:
            weight=bullet[5].text
        except Exception as e:
            weight=None
        book_details.append(weight)
        try:
            dimensions=bullet[6].text
        except Exception as e:
            dimensions=None
        book_details.append(dimensions)
        rev=booksoup.find_all('div', class_='a-section review aok-relative')
        for i in range(0,3):
            try:
                review=rev[i].find_all('span', class_='')
                review=review[1].text
            except Exception as e:
                review=None
            book_details.append(review)
        
        table.append(book_details)
    return table
```

```python
import pandas as pd 
page=[]
for i in range(1,3):
    page.append(get_data(i))
flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(flatten(page),columns=['Publisher','Language','Paperback','ISBN-10', 'ISBN-13','Weight','Dimentions', 'Review-1', 'Review-2', 'Review-3'])
```
